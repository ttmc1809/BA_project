```python
# Import Library and Modle
import numpy as np
import pandas as pd
import seaborn as sns
import statistics as st
import matplotlib.pyplot as plt
from scipy.stats import skew
from scipy.stats import kurtosis
from scipy.optimize import curve_fit
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import LogisticRegression
from mpl_toolkits.mplot3d import Axes3D
import yfinance as yf
from statsmodels.tsa.arima.model import ARIMA
```

<h1>I. Descriptive Statistics</h1>

<h2>1. Statistic</h2>

```python
# Import Dataset
 # EUR - VND
eur = pd.read_csv("EUR_VND.csv")
 # GBP - VND
gbp = pd.read_csv("GBP_VND.csv")
 # JPY - VND
jpy = pd.read_csv("JPY_VND.csv")

# Convert
eur = eur['Sale'].replace({',':''}, regex=True).astype(float)
gbp = gbp['Sale'].replace({',':''}, regex=True).astype(float)
jpy = jpy['Sale'].replace({',':''}, regex=True).astype(float)
```

```python
# Statistic
    # Count
ecount = len(eur)
gcount = len(gbp)
jcount = len(jpy)
    # Mean
emean = st.mean(eur)
gmean = st.mean(gbp)
jmean = st.mean(jpy)
    # Standard deviation
estd = st.stdev(eur)
gstd = st.stdev(gbp)
jstd = st.stdev(jpy)
    # Min
emin = min(eur)
gmin = min(gbp)
jmin = min(jpy)
    # Quantile (0.25)
e25 = np.quantile(eur,0.25)
g25 = np.quantile(gbp,0.25)
j25 = np.quantile(jpy,0.25)
    # Quantile (0.50)
e5 = np.quantile(eur,0.5)
g5 = np.quantile(gbp,0.5)
j5 = np.quantile(jpy,0.5)
    # Quantile (0.75)
e75 = np.quantile(eur,0.75)
g75 = np.quantile(gbp,0.75)
j75 = np.quantile(jpy,0.75)
    # Max
emax = max(eur)
gmax = max(gbp)
jmax = max(jpy)

# Print
print('      | EUR-VND | GBP-VND | JPY-VND')
print(f'Count | {ecount}    | {gcount}    | {jcount}')
print(f'Mean  | {emean:.1f} | {gmean:.1f} | {jmean:.1f}')
print(f'Std   | {estd:.2f} | {gstd:.2f} | {jstd:.2f}')
print(f'Min   | {emin} | {gmin} | {jmin}')
print(f'25%   | {e25} | {g25} | {j25}')
print(f'50%   | {e5} | {g5} | {j5}')
print(f'75%   | {e75} | {g75} | {j75}')
print(f'Max   | {emax} | {gmax} | {jmax}')
```
